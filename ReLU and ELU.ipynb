```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def ReLU(x) -> float:
    return max(x, 0.0)
```

```python
def ReLU_prime(x) -> float:
    return 1. if x > 0 else 0.
```

```python
def ELU(x) -> float:
    if x > 0:
        return x
    else:
        return math.exp(x) - 1.
```

```python
def ELU_prime(x) -> float:
    if x > 0:
        return 1.
    else:
        return math.exp(x)
```

```python
x = np.linspace(-2, 2, num=1000)
relu_x = np.vectorize(ReLU)(x)
relu_prime_x = np.vectorize(ReLU_prime)(x)
elu_x = np.vectorize(ELU)(x)
elu_prime_x = np.vectorize(ELU_prime)(x)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(15, 5))
axs[0].grid(linestyle='-', linewidth=0.5)
axs[0].plot(x, relu_x, label='ReLU', color='red')
axs[0].plot(x, relu_prime_x, label="ReLU'", color='blue')
axs[0].legend()
axs[1].grid(linestyle='-', linewidth=0.5)
axs[1].plot(x, elu_x, label='ELU', color='red')
axs[1].plot(x, elu_prime_x, label="ELU'", color='blue')
axs[1].legend()

plt.show()
fig.savefig("figure.pdf", bbox_inches='tight')
```
